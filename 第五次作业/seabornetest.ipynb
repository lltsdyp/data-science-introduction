```python
# 导入必要的库
import seaborn as sns
import matplotlib.pyplot as plt

# 加载内置的数据集 "tips"
tips = sns.load_dataset("tips")

# 设置绘图风格
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
lineplot = sns.lineplot(x="day", y="total_bill", data=tips)
lineplot.set_title('Total Bill Over Days')
plt.xlabel('Day of Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x="day", y="total_bill", data=tips)
barplot.set_title('Average Total Bill by Day')
plt.xlabel('Day of Week')
plt.ylabel('Average Total Bill')
plt.show()
```

```python
day_totals = tips.groupby('day')['total_bill'].mean().reset_index()
labels = day_totals['day']
sizes = day_totals['total_bill']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Average Total Bill Distribution by Day')
plt.axis('equal')  
plt.show()
```
